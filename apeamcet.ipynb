```python
eval_College = df
eval_College.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
eval_College['Gender'] = [0 if x == 'boys' else 1 for x in eval_College['Gender']]
eval_College['Gender'].unique()
eval_College['Category']= le.fit_transform(eval_College['Category'])
eval_College['Category'].unique()
```

```python
y = eval_College[['College_Name', 'Branch']]
X = eval_College[['Category','Gender', 'Rank']]
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```
